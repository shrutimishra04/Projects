```python
import pandas as pd
```

```python
df=pd.read_csv("salary_predict_dataset.csv")
df.head() #for displaying dataset
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df.duplicated().sum()
```

```python
x=df.iloc[:,:4] #removed salary column
x.head()
```

```python
#y=df.loc[:,['Salary']] #selecting the salary column by label
y=df.iloc[:,4:] #selecting by position
y
```

split the data in two parts , one for training and other for testing. 

```python
from sklearn.model_selection  import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size =0.2,random_state=0) #random_state is used for  same accuracy
```

```python
x_test
```

```python
y_test #original value
```

```python
from sklearn.linear_model import LinearRegression,LogisticRegression    #all the ml algos will be present in sklearn
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score


lr=LinearRegression() #creating object  #on continuous value we use regression and on discreate value we use classification
lr1=LogisticRegression()
dt=DecisionTreeRegressor()
rf=RandomForestRegressor()


lr.fit(x_train,y_train) #call fit method and pass two variable
lr1.fit(x_train,y_train)
dt.fit(x_train,y_train)
rf.fit(x_train,y_train)


y_pred1=lr.predict(x_test) #call predict method used double [[]] because data is in form of rows and column
y_pred2=lr1.predict(x_test)
y_pred3=dt.predict(x_test)
y_pred4=rf.predict(x_test)


print(f"Predicted salary for Linear Regression,{y_pred1} & accuracy is {r2_score(y_test,y_pred1)}")  #r2_score is difference original value and predicted vale it shows the accuracy 
print(f"Predicted salary for Logistic Regression,{y_pred2} & accuracy is {r2_score(y_test,y_pred2)}")
print(f"Predicted salary for Decision Tree,{y_pred3} & accuracy is {r2_score(y_test,y_pred3)}")
print(f"Predicted salary for Random Forest,{y_pred4} & accuracy is {r2_score(y_test,y_pred4)}")
```
